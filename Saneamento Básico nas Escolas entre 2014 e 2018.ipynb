## 2. Criando listas e variáveis

```python
#Criando listas e variáveis.
Dados2014 = []
Dados2015 = []
Dados2016 = []
Dados2017 = []
Dados2018 = []
anos= ['2014', '2015', '2016', '2017', '2018']
Dados = [Dados2014, Dados2015, Dados2016, Dados2017, Dados2018]
```

## 3. Quantidade de escolas com acesso a *ÁGUA* 

### 3.1 Importação dos Dados

```python
Dados2014.append(df2014.query('(DESC_SITUACAO_FUNCIONAMENTO == 1) and (ID_AGUA_INEXISTENTE == 0)')['DESC_SITUACAO_FUNCIONAMENTO'].count())
Dados2015.append(df2015.query('(TP_SITUACAO_FUNCIONAMENTO == 1) and (IN_AGUA_INEXISTENTE == 0)')['TP_SITUACAO_FUNCIONAMENTO'].count())
Dados2016.append(df2016.query('(TP_SITUACAO_FUNCIONAMENTO == 1) and (IN_AGUA_INEXISTENTE == 0)')['TP_SITUACAO_FUNCIONAMENTO'].count())
Dados2017.append(df2017.query('(TP_SITUACAO_FUNCIONAMENTO == 1) and (IN_AGUA_INEXISTENTE == 0)')['TP_SITUACAO_FUNCIONAMENTO'].count())
Dados2018.append(df2018.query('(TP_SITUACAO_FUNCIONAMENTO == 1) and (IN_AGUA_INEXISTENTE == 0)')['TP_SITUACAO_FUNCIONAMENTO'].count())
```

### 3.2 Valores de cada ano

```python
print (f'Os dados de 2018 são: {Dados2018}')
print (f'Os dados de 2017 são: {Dados2017}')
print (f'Os dados de 2016 são: {Dados2016}')
print (f'Os dados de 2015 são: {Dados2015}')
print (f'Os dados de 2014 são: {Dados2014}')
```

### 3.3 Representação Gráfica

```python
fig,ax = plt.subplots()
plt.rcParams['figure.figsize'] = (10,8)
plt.plot(anos, Dados, color='black')
plt.scatter(anos, Dados, color='blue')
ax.plot(anos, Dados, label='Evolução')
ax.grid()
plt.legend()
plt.title('Registros da quantidade de escolas entre 2014 e 2018 com ao acesso a água.')
plt.ylabel('Quantidade de Escolas')
plt.xlabel('Anos')
plt.show(Dados)
```
